```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import requests

# Extraer datos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Restablecer Ã­ndice
tesla_data.reset_index(inplace=True)

# Mostrar primeras 5 filas
tesla_data.head()
```

```python
# Usar yfinance para obtener ingresos de Tesla
tesla_revenue = tesla.financials.T  # Transpuesta para ver aÃ±os como filas

# Mostrar Ãºltimas 5 filas
tesla_revenue.tail()
```

```python
# Extraer datos de GameStop
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Restablecer Ã­ndice
gme_data.reset_index(inplace=True)

# Mostrar primeras 5 filas
gme_data.head()
```

```python
# Usar yfinance para obtener ingresos de GameStop
gme_revenue = gme.financials.T

# Mostrar Ãºltimas 5 filas
gme_revenue.tail()
```

```python
def make_graph(stock_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(stock_data['Date'], stock_data['Close'], label="Precio de Cierre")
    plt.xlabel("Fecha")
    plt.ylabel("Precio (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
make_graph(tesla_data, "EvoluciÃ³n del precio de las acciones de Tesla (TSLA)")
```

```python
make_graph(gme_data, "EvoluciÃ³n del precio de las acciones de GameStop (GME)")
```
